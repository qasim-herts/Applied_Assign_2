```python
#Need indicator IDS to import data from World bank in DataFrame
#used web address method example (https://data.worldbank.org/indicator/EG.ELC.ACCS.ZS)
indicator_ids_1=['EG.ELC.ACCS.ZS','SP.POP.GROW','SP.URB.TOTL']
country_codes= ['AFG','ALB','DZA','ASM','AND']

df_indicator_ids_1 = wb.data.DataFrame(indicator_ids_1,country_codes,range(2014,2018))
ad=pd.DataFrame(df_indicator_ids_1)
ad.to_csv('web_data.csv') #data is imported in dataframe from world bank and store in CVS formate
print(ad)
```

```python
#Funcation which the file name as argument and return two dataframes 
def ingest_manipulate(file_name=r'D:\New folder\a_Data.csv'):
    df = pd.read_csv(file_name)
    df.drop(['Country Name', 'Series Name'], axis=1, inplace=True)
    df1 = df.groupby(['Country Code', 'Series Code']).sum()
    return df1, df1.T
```

```python
#funcation cal 
df1, df2 = ingest_manipulate()
df1.to_csv('Dataframe1.csv')
df2.to_csv('Dataframe2.csv')
```

```python
#visualization of population growth annual 
N = 5
ind = np.arange(N) 
width = 0.25
plt.rcParams["figure.figsize"] = (10,6)
x = [3.35560233082199,-0.207046999760594,2.03358715504452,0.132725922558567,-1.94651826894114]
bar1 = plt.bar(ind, x, width, color = 'r')
  
y = [3.07705362914441,-0.291205786840436, 2.0453720881528,0.026882443238318,-1.55213476921424]
bar2 = plt.bar(ind+width, y, width, color='g')
  
z = [2.77803478841236,-0.159880412127734, 2.05135237337878,-0.120130903326101,-0.898980862435393]
bar3 = plt.bar(ind+width*2, z, width, color = 'b')

  
p = [2.54783303994568, -0.0919722937442495, 2.04490931275225,-0.219117153655893,-0.38628103925755]
bar4 = plt.bar(ind+width/2, p, width)


plt.xlabel("country")
plt.ylabel('Percentage')
plt.title("Population growth (annual %))")
  
plt.xticks(ind+width,['Afghanistan', 'Albania', 'Algeria','American Samoa','Andorra'])
plt.legend( (bar1, bar2, bar3, bar4), ('2014', '2015', '2016','2017') )
plt.show()
```

```python
#visualization of population growth annual with rapid increasing of urban population
plt.rcParams["figure.figsize"] = (10,6)
data = {'Population growth (annual %)': [3.35560233082199, 3.07705362914441,2.77803478841236, 2.54783303994568],
        'Urban population(% of Population)': [8.204880, 8.535606, 8.852834, 9.164768] }
data_frame = pd.DataFrame(data,columns=['Population growth (annual %)','Urban population(% of Population)'], index = ['2014','2015','2016','2017',])

data_frame.plot.barh()
plt.show()
```
